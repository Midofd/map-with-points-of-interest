```python
import geopandas
import folium
```

```python
# Method to convert geopoint to map plt
def json_to_map(JSON_files_path=[]):
    # Make loop on all files to parse the files to dict
    for file_path in JSON_files_path:
        # Read the JSON data by geopandas
        df =  geopandas.read_file(file_path)

        # Set the first point as a location 
        if file_path==JSON_files_path[0]:
            m = folium.Map(location= [df['geometry'][0].y, df['geometry'][0].x], zoom_start=14,control_scale=True)

        # get all the point from file
        for point in df['geometry']:
            p_long=point['geometry'].y
            p_lat=point['geometry'].x
            place_name=point['properties']['title']
            folium.Marker(
            [p_long, p_lat], popup=f"<b>{place_name}</b>", tooltip=place_name[:3]+"...",
            icon=folium.Icon(color='blue',icon_color='white',icon='ok-circle')
        ).add_to(m)
    # Print the plot
    return m
```
