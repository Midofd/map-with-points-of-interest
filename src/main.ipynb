```python
import geopandas
import folium
from folium.plugins import MousePosition,AntPath
import ipywidgets as widgets
from geopy import distance
import  IPython.display as out
```

```python
# Method to convert geopoint to map plt
def json_to_map(JSON_files_path=[]):
    #liste to store all cites list to display him in dropdown list
    points_name=[]

    # Make loop on all files to parse the files to dict
    for file_path in JSON_files_path:
        # Read the JSON data by geopandas
        df =  geopandas.read_file(file_path)

        # Set the first point as a location 
        if file_path==JSON_files_path[0]:
            m = folium.Map(location= [df['geometry'][0].y, df['geometry'][0].x], zoom_start=15,control_scale=True)
            
        folium.TileLayer(
        tiles='https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
        attr='Esri',
        name='Esri Satellite',
        overlay=False,
        control=True,detect_retina=True).add_to(m)

        # get all the point from file
        for point in df.iterrows():
            p_long=point[1][1].y
            p_lat=point[1][1].x
            place_name=point[1][0]
            points_name.append([place_name,p_long,p_lat])
            folium.Marker(
            [p_long, p_lat], popup=f"<b>{place_name}</b>", tooltip=place_name[:3]+"...",
            icon=folium.Icon(color='blue',icon_color='white',icon='ok-circle')
        ).add_to(m)
    MousePosition().add_to(m)
    folium.LatLngPopup().add_to(m)
    # Print the plot
    return m,points_name
```
